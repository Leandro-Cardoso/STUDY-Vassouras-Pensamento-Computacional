3) Você fez duas provas. **p1 e p2.** Você precisa descobrir qual a sua média.
Crie um programa o qual você insere a sua nota e ao final, irá fornecer o resultado dessa média.

````python
p1 = float(input('Qual sua nota na P1? '))
p2 = float(input('Qual sua nota na P2? '))

media = (p1 + p2) / 2

print(f'Sua média é {media:.1f}')
````

4) Crie um programa que calcule seu IMC (indice de massa corpórea). Lembrando que para elevar um número ao quadrado, usa o **
```
imc = peso / altura²

```

````python
peso = float(input('Qual o seu peso em kg? '))
altura = float(input('Qual a sua altura em metros? '))

imc = peso / altura ** 2

print(f'O seu IMC é {imc:.1f}')
````

 7) Crie um programa que pergunte o nome de um produto e o preço.
 Ao final ele deve responder o nome e o preço dele com um acréscimo de 30%.

````python
produto = input('Nome do produto: ')
preco = float(input(f'Preço do(a) {produto}: '))
acrescimoPorcentagem = .3

precoFinal = preco * (1 + acrescimoPorcentagem)

print(f'O preço final do(a) {produto} é de {precoFinal:.2f}')
````

10) Crie um programa que após fornecer um valor total de uma conta de restaurante, ele adiciona uma taxa de 15% de aumento na conta. E faça a divisão pelo número de pessoas na mesa



````python
valorConta = float(input('Valor da conta: '))
numeroPessoas = int(input('Quantidade de pessoas: '))
taxaPorcentagem = .15

valorTotalFinal = valorConta * (1 + taxaPorcentagem)
valorPorPessoa = valorTotalFinal / numeroPessoas

print(f'O valor a ser pago por pessoa é {valorPorPessoa:.2f}')
````
